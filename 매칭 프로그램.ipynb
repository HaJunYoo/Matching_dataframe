```python
def encode_rank(x):
    # Convert to down if trade change is less than zero
    if x == 'A':
        return 5
    # Convert to up if trade change is greater than zero
    elif x == 'B':
        return 4
    # Convert to stable if trade change is zero
    elif x == 'C':
        return 3
    # Convert to stable if trade change is zero
    elif x == 'D':
        return 2
    # Convert to stable if trade change is zero
    elif x == 'F':
        return 1
    # Convert to stable if trade change is zero
    else :
        return 0
```

```python
df2['등급'] = df2.loc[:,['등급']].applymap(encode_rank)
```
